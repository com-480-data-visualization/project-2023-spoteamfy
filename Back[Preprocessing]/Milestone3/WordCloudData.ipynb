# Retrieve tags and their occurences for WordCloud 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/tracks_final.csv')

#Transforming the tags column into a list of tags
df['tags'] = df.tags.apply(lambda x : np.array(x.replace('[', '')
                                             .replace(']', '')
                                             .replace('\'', '')
                                             .replace(',', '')
                                             .split()))
#Get one row per tag (multiple rows for one song)
df = df.explode('tags')

tags_per_emotion = df.groupby("emotions").agg({'tags': lambda x: list(x)})

tags_per_emotion_dict = tags_per_emotion.to_dict()['tags']
```

```python
for key, value in tags_per_emotion_dict.items():
    #Count occurence of each tag in value
    tags_per_emotion_dict[key] = {i:value.count(i) for i in value}

for key, value in tags_per_emotion_dict.items():
    #Sort the dict by value
    tags_per_emotion_dict[key] = [{'word' : k, 'size' : v} for k, v in value.items()]
tags_per_emotion_dict
```

```python
#Write the dict to JSON
import json
with open('data/tags_per_emotion.json', 'w') as fp:
    json.dump(tags_per_emotion_dict, fp)
```
