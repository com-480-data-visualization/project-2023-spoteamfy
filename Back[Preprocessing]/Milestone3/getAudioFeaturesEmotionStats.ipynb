```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tracks_final.csv')
df
```

```python
to_keep = ['emotions', 'danceability', 'arousal', 'dominance', 'energy', 'loudness',
             'acousticness', 'instrumentalness', 'valence']
df = df[to_keep]
df
```

```python
#min max normalization
my_df = df.copy()

for i in to_keep[1:]:
    my_df[i] = round((df[i] - df[i].min()) / (df[i].max() - df[i].min()) * 100).astype(int)

my_df
```

```python
#bootstrap function 
def bootstrap_IC(serie, n = 1000):
    
    row = {}
    means = []
    
    for _ in np.arange(n):
        sampling = serie.sample(n=serie.size, replace=True)
        means.append(float(sampling.mean()))
        
    means.sort()
    
    row['IC_low']  = np.percentile(means, 2.5).round(decimals=1)
    row['mean']    = np.percentile(means, 50).round(decimals=1)
    row['IC_high'] = np.percentile(means, 97.5).round(decimals=1)
    
    return pd.Series(row)
```

```python
features = list(my_df.columns)[1:]

dict_features = {}

for feature in features : 
    
    #current processed feature
    print(feature)
    
    #do the computation
    feature_df   = my_df[['emotions', feature]].groupby(['emotions']).apply(bootstrap_IC).sort_values(by=['mean'], ascending=False)
    
    #convert into dictionary
    feature_dico = feature_df.to_dict('index')
    
    dict_features[feature] = feature_dico 
```

```python
import json

#save .json file
with open('data/features_stat_by_emotions.json', 'w') as fp:
    json.dump(dict_features, fp)
```
